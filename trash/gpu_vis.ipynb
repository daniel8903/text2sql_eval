```python
import pandas as pd
import matplotlib.pyplot as plt

CSV_FILE = "gpu_metrics_log_phi.csv"

# Define your GPU's hardware limits manually
GPU_LIMITS = {
    "temperature_c": 90,          # Example: 90°C safe upper limit
    "gpu_util_percent": 100,
    "vram_used_mb": 12282,
    "power_w": 215,
    "clock_gpu_mhz": 3000,        # Adjust based on your 4070's boost clock
    "fan_speed_percent": 100,
}

def load_data():
    df = pd.read_csv(CSV_FILE, parse_dates=["timestamp"])
    return df

def plot_metric(df, column, ylabel, title, color):
    plt.figure(figsize=(12, 4))
    plt.plot(df["timestamp"], df[column], label=column, color=color)

    # Use fixed limit if available
    if column in GPU_LIMITS:
        plt.ylim(0, GPU_LIMITS[column])
        plt.title(f"{title} (max scale: {GPU_LIMITS[column]})")
    else:
        plt.title(title)

    plt.ylabel(ylabel)
    plt.xlabel("Zeit")
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def main():
    df = load_data()
    df = df[df["gpu_index"] == 0]

    plot_metric(df, "temperature_c", "°C", "GPU-Temperatur", "red")
    plot_metric(df, "gpu_util_percent", "%", "GPU-Auslastung", "orange")
    plot_metric(df, "vram_used_mb", "MB", "VRAM-Nutzung", "blue")
    plot_metric(df, "power_w", "Watt", "Stromverbrauch", "green")
    plot_metric(df, "clock_gpu_mhz", "MHz", "GPU-Taktfrequenz", "purple")
    plot_metric(df, "fan_speed_percent", "%", "Lüftergeschwindigkeit", "cyan")

if __name__ == "__main__":
    main()
```
